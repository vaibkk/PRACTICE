# Create groups of companies based on market cap


```python
# create condition for company_type

conditions =  [
    (nse_data['marketcap'] <= 5000),
    (nse_data['marketcap'] > 5000) & (nse_data['marketcap'] <= 20000),
    (nse_data['marketcap'] > 20000), 
    (nse_data['marketcap'].isna()) # for companies with invalid marketcap values
        ]

# Create company type values
cap_values = ['small-cap', 'mid-cap', 'large-cap', 'invalid']

# Create a new column company_type in nse_data
nse_data['company_type'] = np.select(conditions, cap_values)

# View the details of company, marketcap, & company_type
nse_data[{'company', 'marketcap', 'company_type'}]
```

```python
# joins
band_members = pd.DataFrame({"name": ["John", "Paul", "Mick"], 
"band": ["Beatles", "Beatles", "Stones"], "likes": ["food", "baseball", "Texas"]})
print("band_members :\n", band_members)

band_instruments = pd.DataFrame({"name": ["John", "Paul", "Keith"], 
 "plays" : ["guitar", "bass", "guitar"], "food": ["Mexican", "Italian", "Pizza"]})

print("band_instruments : \n", band_instruments)
```

```python
# left join
band_instruments.merge(band_members, how = "left", on = "name")
```

```python
# left join
band_members.merge(band_instruments, how = "left", on = "name")
```

```python
# right join
band_members.merge(band_instruments, how = "right", on = "name")
```

```python
# inner join
band_members.merge(band_instruments, how = "inner", on = "name")
```

```python
# outer join
band_members.merge(band_instruments, how = "outer", on = "name")
```

```python
# outer join
band_members.merge(band_instruments, how = "outer", on = "name")
```

```python
# outer join
band_members.merge(band_instruments, how = "outer",  on = "name",  indicator = True)
```

```python
# outer join
band_members.merge(band_instruments, how = "outer",  left_on = "name", right_on = "name", indicator = True)
```

```python
# joins
band_members = pd.DataFrame({"bm_name": ["John", "Paul", "Mick"], 
"band": ["Beatles", "Beatles", "Stones"], "likes": ["food", "baseball", "Texas"]})
print("band_members :\n", band_members)

band_instruments = pd.DataFrame({"bi_name": ["John", "Paul", "Keith"], 
 "plays" : ["guitar", "bass", "guitar"], "food": ["Mexican", "Italian", "Pizza"]})

print("band_instruments : \n", band_instruments)
```

```python
# outer join
band_members.merge(band_instruments, how = "outer",  left_on = "bm_name", right_on = "bi_name", indicator = True)
```

# Filling operations on time series data

```python
# create a series type dataset with time features

test_ts = pd.Series({"date" : ['20211130', '20211129', '20211128', '20211126', '20211125', '20211124'], 
"temp": [7, 6, 6.5, '', 8, 7.5], "humidity": [42, 80, '', 54, '', 74]} )
```

```python
# create a series type dataset with time features
date = pd.date_range(start = '25/11/2021', end = '30/11/2021', freq = 'D')
# use np.NaN to create NaN values
night_temp = [7, 6, 6.5, np.NaN, 8, 7.5]
humidity = [42, 80, '', 54, np.NaN, 74]
temp_df = pd.Series(night_temp, index= date )
```

```python
# verify type of test_ts
temp_df
```

```python
# perform forward fill
# forward fill uses the prior value
temp_df.resample('1D').mean().ffill()
```

```python
# perform backward fill
# backward fill uses the next value
temp_df.resample('1D').mean().bfill()
```

```python
# perform interpolate fill
# interpolate fill replaces NaN based on interpolation of non-NaN values
temp_df.resample('1D').mean().interpolate()
```
